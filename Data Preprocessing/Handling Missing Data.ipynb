# Deals with Missing Values

```python
import pandas as pd 
import numpy as np
```

```python
dataset = pd.read_csv("Dataset/AB_NYC_2019.csv")
df = dataset.copy()
```

```python
df.isna().sum()
```

## Deleting the Column with missing data

```python
df1 = df.copy()
```

```python
df1.head()
```

```python
df1.isna().sum()
```

```python
df1.drop("last_review",axis=1,inplace=True)
```

```python
df1.isna().sum()
```

## Deleting the row with missing data

```python
df2 = df.copy()
```

```python
df2.isna().sum()
```

```python
df2.dropna(inplace=True)
```

```python
df2.isna().sum()
```

```python
df2.shape
```

## Filling the Missing Values - Imputation
   (i) Numerical Data - use Mean

```python
df3 = df.copy()
```

```python
df3.head()
```

```python
mean_value = df3["reviews_per_month"].mean()
mean_value
```

```python
df3["reviews_per_month"].fillna(mean_value,inplace=True)
df3.isna().sum()
```

## FIlling Categorial Data

```python
df3.nunique()
```

```python
df3["last_review"].fillna(df3["last_review"].value_counts().index[0],inplace=True)
```

```python
df3.isna().sum()
```

### FIling Categorial data with new category

```python
df4 = df.copy()
```

```python
df4.isna().sum()
```

```python
df4["last_review"].fillna("Not Reviewed",inplace=True)
df4.sample(10)
```

## Advance Imputation

```python
df5 = df.copy()
df6 = df.copy()
df7 = df.copy()
```

```python
df5.isna().sum()
```

```python
df5["reviews_per_month"].interpolate(inplace=True)
df5
```

## Interpolate - Linear

```python
df6.isna().sum()
```

```python
df6["reviews_per_month"].interpolate(method='linear' ,inplace=True)
df6
```

## Interpolate - Quadratic

```python
df7["reviews_per_month"].interpolate(method='quadratic' ,inplace=True)
df7
```
